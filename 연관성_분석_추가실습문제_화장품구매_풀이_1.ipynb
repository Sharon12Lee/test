\# Association Rules 추가 실습

```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cosine
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

# Cosmetics Purchases 아이템 간 연관성 분석

```python
cosmetics_df = pd.read_csv('Cosmetics.csv')
cosmetics_df.set_index('Trans. ',inplace=True)
cosmetics_df.head()
```

```python
frequentItemsets = apriori(cosmetics_df, min_support=0.1, use_colnames=True)
rules = association_rules(frequentItemsets)
results = rules.sort_values(by=['lift'], ascending=False)
results.drop(columns=['antecedent support', 'consequent support', 'conviction', 'zhangs_metric'])
```
